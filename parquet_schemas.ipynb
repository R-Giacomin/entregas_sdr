# Parquet Schema Explorer
This notebook lists all `.parquet` files in the current directory and displays their schemas and a sample of the data.

```python
import pandas as pd
import glob
import os
from IPython.display import display, Markdown
```

```python
# Find all parquet files
parquet_files = glob.glob("*.parquet")

if not parquet_files:
    print("No Parquet files found.")
else:
    for file in parquet_files:
        display(Markdown(f"## File: `{file}`"))
        
        # Read the file
        df = pd.read_parquet(file)
        
        display(Markdown("### Schema (dtypes)"))
        print(df.dtypes)
        
        display(Markdown("### Data Preview (head)"))
        display(df.head())
        
        display(Markdown("---"))
```
